```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/myvalume/module03/data/Test_knight.csv')
df
```

```python
df.iloc[:, 0].describe()
```

```python
# keep first column in the dataframe
f = df.iloc[:, 0]
# Plotting histogram the first column
plt.hist(f, color='blue', bins=40, alpha=0.7, histtype='barstacked')
plt.title(f.name)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Plotting histogram for all columns in the dataframe
k = 0
fig, ax = plt.subplots(6, 5, figsize=(24, 24))
for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        if k < len(df.columns):
            f = df.iloc[:, k]
            ax[i, j].hist(f, color='green', bins=40, alpha=0.5, histtype='barstacked')
            ax[i, j].set_title(f.name)
            ax[i, j].grid(False)
            ax[i, j].legend(['Knight'])
            k += 1
plt.show()
```

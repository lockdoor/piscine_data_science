```python
# read the CSV file
test_knight = pd.read_csv('/myvolume/subject/data/Test_knight.csv')
train_knight = pd.read_csv('/myvolume/subject/data/Train_knight.csv')
```

```python
# make cmap by train_knight['knight'] it store string 'Jedi' or 'Sith'
cmap = train_knight['knight'].map({'Jedi': 'blue', 'Sith': 'red'})
```

```python
# Ploting scatter plot with most correlation of train_knight
plt.scatter(train_knight['Empowered'], train_knight['Prescience'], c=cmap, alpha=0.5,)
plt.title('Train Separation')
plt.xlabel('Empowered')
plt.ylabel('Prescience')
plt.scatter([], [], c='blue', label='Jedi')
plt.scatter([], [], c='red', label='Sith')
plt.legend()
plt.show()
```

```python
# plotting scatter plot with less correlation of train_knight
plt.scatter(train_knight['Midi-chlorien'], train_knight['Push'], c=cmap, alpha=0.5)
plt.title('Train Mixed')
plt.xlabel('Midi-chlorien')
plt.ylabel('Push')
plt.scatter([], [], c='blue', label='Jedi')
plt.scatter([], [], c='red', label='Sith')
plt.legend()
plt.show()
```

```python
# Ploting scatter plot with most correlation of test_knight
plt.scatter(test_knight['Empowered'], test_knight['Prescience'], alpha=0.5, label='Knight')
plt.title('Test Separation')
plt.xlabel('Empowered')
plt.ylabel('Prescience')
plt.legend()
plt.show()
```

```python
# plotting scatter plot with less correlation of test_knight
plt.scatter(test_knight['Midi-chlorien'], test_knight['Push'], alpha=0.5, label='Knight')
plt.title('Train Mixed')
plt.xlabel('Midi-chlorien')
plt.ylabel('Push')
plt.legend()
plt.show()
```

```python
# Plotting 4 graphs in one figure
def plot_four_charts(one, two, three, four):
    _, axs = plt.subplots(2, 2, figsize=(10, 8))
    axs[0, 0].scatter(train_knight[one], train_knight[two], c=cmap, alpha=0.5)
    axs[0, 0].set_title('Train Separation')
    axs[0, 0].set_xlabel(one)
    axs[0, 0].set_ylabel(two)
    axs[0, 0].scatter([], [], c='blue', label='Jedi')
    axs[0, 0].scatter([], [], c='red', label='Sith')
    axs[0, 0].legend()

    axs[0, 1].scatter(train_knight[three], train_knight[four], c=cmap, alpha=0.5)
    axs[0, 1].set_title('Train Mixed')
    axs[0, 1].set_xlabel(three)
    axs[0, 1].set_ylabel(four)
    axs[0, 1].scatter([], [], c='blue', label='Jedi')
    axs[0, 1].scatter([], [], c='red', label='Sith')
    axs[0, 1].legend()

    axs[1, 0].scatter(test_knight[one], test_knight[two], alpha=0.5, label='Knight')
    axs[1, 0].set_title('Test Separation')
    axs[1, 0].set_xlabel(one)
    axs[1, 0].set_ylabel(two)
    axs[1, 0].legend()

    axs[1, 1].scatter(test_knight[three], test_knight[four], alpha=0.5, label='Knight')
    axs[1, 1].set_title('Test Mixed')
    axs[1, 1].set_xlabel(three)
    axs[1, 1].set_ylabel(four)
    axs[1, 1].legend()
    plt.tight_layout()
    plt.show()

# plot look like the subject it choose Empowered and Stims
plot_four_charts('Empowered', 'Stims', 'Push', 'Deflection')
```

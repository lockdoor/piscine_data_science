```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# read the CSV file
train_knight = pd.read_csv('/myvolume/subject/data/Train_knight.csv')
```

```python
# train_test_split is used to split the DataFrame into two parts
# stratify is used to ensure that the split maintains the same distribution of classes
# in the target variable (knight) in both the training and testing sets

# it split amount of Jedi and Sith is likely originally avoid bias data

training_df, sample_df = train_test_split(
    train_knight, 
    test_size=0.10,             # Keep 10% 
    stratify=train_knight['knight'], 
    random_state=42
)
```

```python
print('shape_train:', train_knight.shape)
print ('shape-sample:', sample_df.shape)
print ('shape-training:', training_df.shape)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 6))
train_knight['knight'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=ax[0])
ax[0].set_title('Population')
sample_df['knight'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=ax[1])
ax[1].set_title('Sample')
training_df['knight'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=ax[2])
ax[2].set_title('Training')
plt.show()
```

```python
# save 3 dataframes to CSV files in the same directory
train_knight.to_csv('/myvolume/module03/ex05/Train_knight.csv', index=False)
sample_df.to_csv('/myvolume/module03/ex05/Validation_knight.csv', index=False)
training_df.to_csv('/myvolume/module03/ex05/Training_knight.csv', index=False)
```

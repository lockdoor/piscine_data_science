```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the data
# train_knight_ds_path = 'myvolume/subject/data04/Train_knight.csv'
train_knight_ds_path = '../../subject/data04/Train_knight.csv'

df = pd.read_csv(train_knight_ds_path)

# add column knight_numeric for cast column to numeric
df['knight_numeric'] = df['knight'].map({'Jedi': 1, 'Sith': 0})

# drop the original knight column
df.drop(columns=['knight'], inplace=True)

# rename knight_numeric to knight
df.rename(columns={'knight_numeric': 'knight'}, inplace=True)
```

```python
# make correlation matrix
# correlation is nubber between -1 and 1
correlation = df.corr()
correlation.head()
```

```python
# Make a Heatmap to see the Correlation Coefficient between the data
plt.figure(figsize=(10,8))
sns.heatmap(correlation, vmax=1, vmin=-1, center=0, cmap='coolwarm')
plt.title('Correlation Coefficient', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.gca().xaxis.tick_bottom()
plt.show()
```
